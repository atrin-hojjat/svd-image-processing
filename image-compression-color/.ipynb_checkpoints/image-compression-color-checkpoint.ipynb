```python
image_name = "cat"
image_path = "./cat.png"
```

```python

from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
import os

image = Image.open(image_path)

# Convert the image to grayscale (black and white)
bw_image_array = np.array(image)

plt.imshow(bw_image_array)
plt.axis('off') 
plt.show()
```

```python
R_U, R_S, R_VT = np.linalg.svd(bw_image_array[:, :, 0])
G_U, G_S, G_VT = np.linalg.svd(bw_image_array[:, :, 1])
B_U, B_S, B_VT = np.linalg.svd(bw_image_array[:, :, 2])
```

```python
def compress_image(U, S, Vt, rank):
    return U[:, :rank] @ np.diag(S[:rank]) @ Vt[:rank]
```

```python
def show_image(image):
    plt.imshow(image)
    plt.axis('off') 
    plt.show()
```

```python
def save_image(image_array, output_path):
    if image_array.dtype != np.uint8:
        image_array = image_array.astype(np.uint8)
    image = Image.fromarray(image_array)
    image.save(output_path)
```
